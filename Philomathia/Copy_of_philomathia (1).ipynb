````python
# Calcul symbolique de d√©riv√©es avec SymPy
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

x = sp.Symbol('x')

# Exemples de fonctions et leurs d√©riv√©es
fonctions = [
    x**2,
    sp.sin(x),
    sp.exp(x),
    x**3 - 2*x**2 + x - 5
]

print("üìê Calcul de d√©riv√©es symboliques:\\n")
for f in fonctions:
    f_prime = sp.diff(f, x)
    print(f"f(x) = {f}")
    print(f"f'(x) = {f_prime}")
    print()

# Application en Machine Learning: Descente de gradient
# Fonction co√ªt (erreur quadratique): J(Œ∏) = (Œ∏ - 5)¬≤ + 3
theta = sp.Symbol('theta')
J = (theta - 5)**2 + 3
J_prime = sp.diff(J, theta)

print("\\nü§ñ Application ML: Descente de gradient")
print(f"Fonction co√ªt: J(Œ∏) = {J}")
print(f"Gradient: J'(Œ∏) = {J_prime}")
print(f"\\nOptimum (d√©riv√©e = 0): Œ∏* = {sp.solve(J_prime, theta)[0]}")

# Visualisation de la descente de gradient
theta_vals = np.linspace(0, 10, 200)
J_vals = (theta_vals - 5)**2 + 3
J_prime_vals = 2 * (theta_vals - 5)

# Simulation de la descente de gradient
theta_initial = 0.5
learning_rate = 0.1
n_iterations = 15

theta_history = [theta_initial]
for i in range(n_iterations):
    gradient = 2 * (theta_history[-1] - 5)
    theta_new = theta_history[-1] - learning_rate * gradient
    theta_history.append(theta_new)

J_history = [(t - 5)**2 + 3 for t in theta_history]

# Visualisation
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# Graphique 1: Fonction et sa d√©riv√©e
axes[0].plot(theta_vals, J_vals, 'b-', linewidth=2, label='J(Œ∏) = (Œ∏-5)¬≤ + 3')
axes[0].plot(theta_vals, J_prime_vals, 'r--', linewidth=2, label="J'(Œ∏) = 2(Œ∏-5)")
axes[0].axhline(0, color='black', linewidth=0.5)
axes[0].axvline(5, color='green', linestyle=':', linewidth=2, label='Minimum (Œ∏=5)')
axes[0].scatter([2, 5, 8], [(2-5)**2+3, 3, (8-5)**2+3], s=100, c='orange', zorder=5)
axes[0].set_xlabel('Œ∏')
axes[0].set_ylabel('Valeur')
axes[0].set_title('Fonction co√ªt et sa d√©riv√©e')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# Graphique 2: Descente de gradient
axes[1].plot(theta_vals, J_vals, 'b-', linewidth=2, alpha=0.5, label='J(Œ∏)')
axes[1].plot(theta_history, J_history, 'ro-', markersize=8, linewidth=2, label='Descente de gradient')
axes[1].scatter([theta_history[0]], [J_history[0]], s=200, c='red', marker='*',
                edgecolors='black', linewidths=2, zorder=5, label='D√©part')
axes[1].scatter([theta_history[-1]], [J_history[-1]], s=200, c='green', marker='*',
                edgecolors='black', linewidths=2, zorder=5, label='Arriv√©e')

# Ajouter des fl√®ches pour montrer la direction
for i in range(0, min(5, len(theta_history)-1)):
    axes[1].annotate('', xy=(theta_history[i+1], J_history[i+1]),
                    xytext=(theta_history[i], J_history[i]),
                    arrowprops=dict(arrowstyle='->', lw=1.5, color='darkred'))

axes[1].set_xlabel('Œ∏ (param√®tre)')
axes[1].set_ylabel('J(Œ∏) (co√ªt)')
axes[1].set_title(f'Optimisation par descente de gradient (Œ±={learning_rate})')
axes[1].legend()
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print(f"\nüéØ R√©sultat de l'optimisation:")
print(f"  D√©part: Œ∏ = {theta_history[0]:.3f}, J = {J_history[0]:.3f}")
print(f"  Arriv√©e: Œ∏ = {theta_history[-1]:.3f}, J = {J_history[-1]:.3f}")
print(f"  Optimum th√©orique: Œ∏ = 5.000, J = 3.000")
print(f"\\nüí° La d√©riv√©e guide l'algorithme vers le minimum!")

# Exemple pratique: D√©riv√©e partielle pour un mod√®le lin√©aire
print("\\n\\nüìà Exemple ML: R√©gression lin√©aire")
print("Mod√®le: y = ax + b")
print("Co√ªt: MSE = 1/n Œ£(y_pred - y_true)¬≤")
print("\\nGradients (d√©riv√©es partielles):")
print("  ‚àÇMSE/‚àÇa = 2/n Œ£(ax + b - y) √ó x")
print("  ‚àÇMSE/‚àÇb = 2/n Œ£(ax + b - y)")
print("\\n‚Üí Ces gradients permettent d'ajuster a et b pour minimiser l'erreur!")
````

## 13. D√©riv√©e et Descente de Gradient

En science des donn√©es et en IA, les d√©riv√©es sont cruciales pour l'**optimisation**, en particulier dans le contexte de l'apprentissage automatique :

*   **Calcul du Gradient :** Pour les fonctions de plusieurs variables (courantes dans les mod√®les d'apprentissage automatique qui ont de nombreux param√®tres), la g√©n√©ralisation de la d√©riv√©e est le **gradient**. Le gradient est un vecteur qui pointe dans la direction de la plus forte augmentation de la fonction.
*   **Descente de Gradient :** C'est l'un des algorithmes d'optimisation les plus fondamentaux. L'id√©e est de trouver le minimum d'une fonction (souvent une fonction de co√ªt ou d'erreur) en suivant la direction oppos√©e au gradient. Si le gradient pointe vers la mont√©e la plus raide, prendre l'oppos√© nous fait descendre vers le minimum.

La capacit√© de calculer des d√©riv√©es (manuellement, symboliquement avec des outils comme SymPy, ou automatiquement avec des frameworks comme TensorFlow/PyTorch) est essentielle pour entra√Æner la plupart des mod√®les d'apprentissage automatique (r√©gression, r√©seaux de neurones, etc.) en minimisant une fonction de co√ªt.

## 12. Th√©or√®me Central Limite

Le **Th√©or√®me Central Limite (TCL)** est un concept fondamental en probabilit√© et en statistique. Il stipule que, sous certaines conditions, la distribution de la somme (ou de la moyenne) d'un grand nombre de variables al√©atoires ind√©pendantes et identiquement distribu√©es (i.i.d.) tend vers une distribution normale (Gaussienne), quelle que soit la distribution d'origine des variables individuelles.

L'importance du TCL en science des donn√©es et en IA est immense :

*   **Inf√©rence Statistique :** Le TCL justifie l'utilisation de la distribution normale pour construire des intervalles de confiance et effectuer des tests d'hypoth√®ses sur les moyennes d'√©chantillons, m√™me si la distribution de la population d'origine n'est pas normale, √† condition que la taille de l'√©chantillon soit suffisamment grande.
*   **Mod√©lisation :** De nombreux mod√®les statistiques et d'apprentissage automatique qui supposent la normalit√© (comme la r√©gression lin√©aire ou certains tests statistiques) peuvent √™tre appliqu√©s √† des donn√©es agr√©g√©es (comme les moyennes) gr√¢ce au TCL.
*   **Comprendre les ph√©nom√®nes :** Le TCL explique pourquoi de nombreux ph√©nom√®nes naturels et sociaux qui r√©sultent de l'accumulation de nombreux petits effets al√©atoires (erreurs de mesure, tailles des individus, etc.) suivent approximativement une distribution normale.
*   **R√©duction de l'incertitude :** En montrant que la moyenne d'un grand √©chantillon est distribu√©e normalement et que sa variance diminue √† mesure que la taille de l'√©chantillon augmente, le TCL nous aide √† quantifier et √† r√©duire l'incertitude dans nos estimations.

En pratique, "une grande taille d'√©chantillon" est souvent consid√©r√©e comme n > 30, bien que cela d√©pende de la distribution d'origine.

````python
import numpy as np
import matplotlib.pyplot as plt

# D√©monstration du Th√©or√®me Central Limite
# On part d'une distribution non-normale (par exemple, une distribution uniforme)
# dont on va √©chantillonner et calculer les moyennes.

# Param√®tres de la simulation
taille_population = 100000 # Taille de notre "population"
taille_echantillon = 30 # Taille des √©chantillons que nous allons pr√©lever
nombre_echantillons = 1000 # Nombre d'√©chantillons √† pr√©lever

# G√©n√©rer une population √† partir d'une distribution non-normale (Uniforme)
population_data = np.random.uniform(low=0, high=10, size=taille_population)

# Afficher l'histogramme de la population d'origine (non-normale)
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.hist(population_data, bins=50, density=True, alpha=0.7, color='orange', edgecolor='black')
plt.xlabel('Valeur')
plt.ylabel('Densit√© de probabilit√©')
plt.title('Distribution de la Population (Uniforme)')
plt.grid(True, alpha=0.3)


# Pr√©lever de nombreux √©chantillons et calculer la moyenne de chaque √©chantillon
moyennes_echantillons = []
for _ in range(nombre_echantillons):
    echantillon = np.random.choice(population_data, size=taille_echantillon, replace=True) # √âchantillonnage avec remplacement
    moyenne_echantillon = np.mean(echantillon)
    moyennes_echantillons.append(moyenne_echantillon)

# Afficher l'histogramme des moyennes d'√©chantillons
plt.subplot(1, 2, 2)
plt.hist(moyennes_echantillons, bins=30, density=True, alpha=0.7, color='teal', edgecolor='black')
plt.xlabel('Moyenne de l\'√©chantillon')
plt.ylabel('Densit√© de probabilit√©')
plt.title(f'Distribution des Moyennes d\'√âchantillons (n={taille_echantillon})')
plt.grid(True, alpha=0.3)

# Superposer une courbe normale th√©orique pour comparaison
from scipy.stats import norm
# La moyenne de la distribution des moyennes d'√©chantillons est la moyenne de la population
mean_of_sample_means = np.mean(moyennes_echantillons)
# L'√©cart-type de la distribution des moyennes d'√©chantillons est l'√©cart-type de la population / sqrt(n)
# On utilise l'√©cart-type de la population th√©orique uniforme: std = (high - low) / sqrt(12)
std_dev_population = (10 - 0) / np.sqrt(12)
std_error = std_dev_population / np.sqrt(taille_echantillon)

xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mean_of_sample_means, std_error)
plt.plot(x, p, 'k', linewidth=2, label='Distribution Normale Th√©orique')
plt.legend()


plt.tight_layout() # Ajuster l'espacement
plt.show() # Afficher les graphiques

print(f"\nMoyenne de la population d'origine : {np.mean(population_data):.3f}")
print(f"√âcart-type de la population d'origine : {np.std(population_data):.3f}")
print(f"\nMoyenne des moyennes d'√©chantillons : {mean_of_sample_means:.3f}")
print(f"√âcart-type des moyennes d'√©chantillons (Erreur Standard) : {np.std(moyennes_echantillons):.3f}")
print(f"Erreur Standard th√©orique : {std_error:.3f}")
````

La hauteur de chaque barre de l'histogramme repr√©sente la fr√©quence (ou la densit√©) des donn√©es dans cet intervalle.

Les histograms sont essentiels en science des donn√©es et en IA pour :

*   **Visualiser la distribution :** Ils permettent de voir rapidement la forme de la distribution des donn√©es (sym√©trique, asym√©trique, multimodale, etc.).

*   **D√©tecter les outliers :** Les valeurs extr√™mes qui se situent loin du corps principal de la distribution peuvent appara√Ætre comme des barres isol√©es aux extr√©mit√©s.
*   **Comprendre la fr√©quence des valeurs :** Ils montrent quels intervalles de valeurs sont les plus fr√©quents dans le jeu de donn√©es.
*   **Choisir le nombre de bins :** Le choix du nombre de bins peut affecter l'apparence de l'histogramme et la fa√ßon dont la distribution est per√ßue. Un nombre trop faible masque les d√©tails, tandis qu'un nombre trop √©lev√© peut rendre l'histogramme trop granulaire.

Bien qu'un histogramme simple ait √©t√© utilis√© pour visualiser la distribution de probabilit√©, cette section explore plus en d√©tail sa construction et son interpr√©tation.

## 10. Diagramme en Bo√Æte (Boxplot)

Le **diagramme en bo√Æte**, ou **boxplot**, est un outil de visualisation statistique qui r√©sume la distribution d'un ensemble de donn√©es en montrant ses quartiles. Il est particuli√®rement utile pour visualiser la dispersion, l'asym√©trie et identifier les valeurs aberrantes (outliers) dans une distribution.

Un boxplot typique se compose de :

*   **La bo√Æte :** Elle s'√©tend du premier quartile (Q1) au troisi√®me quartile (Q3). La longueur de la bo√Æte est l'Intervalle Interquartile (IQR = Q3 - Q1).
*   **La ligne m√©diane :** Une ligne √† l'int√©rieur de la bo√Æte indique la position du deuxi√®me quartile (Q2), c'est-√†-dire la m√©diane.
*   **Les "moustaches" (whiskers) :** Elles s'√©tendent de la bo√Æte jusqu'aux valeurs minimales et maximales qui ne sont pas consid√©r√©es comme des outliers. Typiquement, elles s'√©tendent jusqu'√† 1.5 * IQR au-del√† de Q1 et Q3.
*   **Les outliers :** Les points de donn√©es qui se situent au-del√† des moustaches sont repr√©sent√©s individuellement sous forme de points.

En science des donn√©es et en IA :

*   Les boxplots sont excellents pour une **analyse exploratoire rapide** de la distribution d'une variable.
*   Ils permettent de **comparer facilement les distributions** de plusieurs groupes ou cat√©gories sur un seul graphique.
*   Ils sont largement utilis√©s pour **d√©tecter visuellement les outliers**, ce qui est une √©tape importante du pr√©traitement des donn√©es.
*   La position de la m√©diane par rapport au centre de la bo√Æte et la longueur relative des moustaches donnent une indication de l'**asym√©trie** de la distribution.

````python
%%markdown
## 3. Matrices

Une **matrice** est un tableau rectangulaire de nombres, de symboles ou d'expressions organis√©s en lignes et en colonnes. Elle est d√©finie par ses dimensions, not√©es m x n, o√π m est le nombre de lignes et n est le nombre de colonnes.

Les matrices sont des outils fondamentaux en alg√®bre lin√©aire et jouent un r√¥le crucial en science des donn√©es et en intelligence artificielle pour plusieurs raisons :

*   **Stockage de donn√©es :** Les ensembles de donn√©es sont souvent repr√©sent√©s sous forme de matrices, o√π chaque ligne correspond √† une observation et chaque colonne √† une caract√©ristique (feature).
*   **Transformations lin√©aires :** Les matrices sont utilis√©es pour repr√©senter des transformations lin√©aires dans l'espace vectoriel, comme les rotations, les mises √† l'√©chelle ou les translations. Ces transformations sont au c≈ìur de nombreux algorithmes, notamment en traitement d'images et en graphisme.
*   **Syst√®mes d'√©quations lin√©aires :** Les matrices permettent de repr√©senter et de r√©soudre efficacement des syst√®mes d'√©quations lin√©aires, ce qui est essentiel dans des domaines comme la r√©gression lin√©aire.
*   **Calculs dans les r√©seaux de neurones :** Les op√©rations matricielles, en particulier la multiplication matricielle, sont les op√©rations de base effectu√©es dans les couches des r√©seaux de neurones pour propager les signaux.
*   **R√©duction de dimensionnalit√© :** Des techniques comme l'Analyse en Composantes Principales (ACP) s'appuient fortement sur les op√©rations matricielles (comme la d√©composition en valeurs singuli√®res) pour r√©duire la dimensionnalit√© des donn√©es tout en pr√©servant l'information importante.

Comprendre les op√©rations matricielles (addition, soustraction, multiplication) est indispensable pour travailler avec des donn√©es structur√©es et construire des mod√®les d'apprentissage automatique.
````

````python
import numpy as np
import matplotlib.pyplot as plt

# 5. Create two NumPy matrices
matrix_a = np.array([[1, 2],
                     [3, 4]])

matrix_b = np.array([[5, 6],
                     [7, 8]])

matrix_c = np.array([[1, 2, 3],
                     [4, 5, 6]])

matrix_d = np.array([[7, 8],
                     [9, 10],
                     [11, 12]])

print("Matrice A (2x2):")
print(matrix_a)
print("\nMatrice B (2x2):")
print(matrix_b)
print("\nMatrice C (2x3):")
print(matrix_c)
print("\nMatrice D (3x2):")
print(matrix_d)


# 6. Demonstrate matrix addition and subtraction
print("\nOp√©rations sur les matrices 2x2 (A et B):")
try:
    somme_ab = matrix_a + matrix_b
    print("A + B =")
    print(somme_ab)
except ValueError as e:
    print(f"Erreur lors de l'addition de A et B: {e}")

try:
    difference_ab = matrix_a - matrix_b
    print("A - B =")
    print(difference_ab)
except ValueError as e:
    print(f"Erreur lors de la soustraction de A et B: {e}")

# Addition/Soustraction entre matrices de dimensions incompatibles (devrait g√©n√©rer une erreur)
print("\nOp√©rations sur matrices de dimensions incompatibles (A et C):")
try:
    somme_ac = matrix_a + matrix_c
    print("A + C =")
    print(somme_ac)
except ValueError as e:
    print(f"Erreur attendue lors de l'addition de A et C: {e}")


# 7. Demonstrate matrix multiplication
# Multiplication A @ B (2x2 @ 2x2 -> 2x2)
print("\nMultiplication matricielle:")
try:
    produit_ab = matrix_a @ matrix_b
    # Alternative: np.dot(matrix_a, matrix_b)
    print("A @ B =")
    print(produit_ab)
except ValueError as e:
    print(f"Erreur lors de la multiplication A @ B: {e}")

# Multiplication C @ D (2x3 @ 3x2 -> 2x2)
try:
    produit_cd = matrix_c @ matrix_d
    print("C @ D =")
    print(produit_cd)
except ValueError as e:
     print(f"Erreur lors de la multiplication C @ D: {e}")

# Multiplication D @ C (3x2 @ 2x3 -> 3x3)
try:
    produit_dc = matrix_d @ matrix_c
    print("D @ C =")
    print(produit_dc)
except ValueError as e:
    print(f"Erreur lors de la multiplication D @ C: {e}")

# Multiplication A @ C (2x2 @ 2x3) - Compatible
try:
    produit_ac = matrix_a @ matrix_c
    print("A @ C =")
    print(produit_ac)
except ValueError as e:
    print(f"Erreur lors de la multiplication A @ C: {e}")

# Multiplication C @ A (2x3 @ 2x2) - Incompatible (nombre de colonnes de C != nombre de lignes de A)
print("\nMultiplication sur matrices de dimensions incompatibles (C @ A):")
try:
    produit_ca = matrix_c @ matrix_a
    print("C @ A =")
    print(produit_ca)
except ValueError as e:
    print(f"Erreur attendue lors de la multiplication C @ A: {e}")

# 8. Demonstrate scalar multiplication
scalaire = 3
mult_scalaire_a = scalaire * matrix_a
print(f"\nMultiplication scalaire ({scalaire} * A):")
print(mult_scalaire_a)


# 9. Generate random data and plot transformation
# G√©n√©rer des points al√©atoires 2D
np.random.seed(42) # pour la reproductibilit√©
points_originaux = np.random.rand(10, 2) * 5 # 10 points, coordonn√©es entre 0 et 5

# Matrice de transformation (par exemple, une rotation + mise √† l'√©chelle)
matrice_transformation = np.array([[1.5, 1],
                                  [-0.5, 1.5]])

# Appliquer la transformation matricielle √† chaque point
# Pour multiplier une matrice (n_points x 2) par une matrice (2x2),
# on peut utiliser @. NumPy g√®re correctement la multiplication ligne par ligne.
points_transformes = points_originaux @ matrice_transformation


# Cr√©er le graphique
plt.figure(figsize=(8, 8))
plt.scatter(points_originaux[:, 0], points_originaux[:, 1], color='blue', label='Points Originaux')
plt.scatter(points_transformes[:, 0], points_transformes[:, 1], color='red', label='Points Transform√©s')

# Ajouter des fl√®ches pour montrer la transformation de quelques points
for i in range(3): # Afficher les transformations pour les 3 premiers points
    plt.annotate('', xy=(points_transformes[i, 0], points_transformes[i, 1]),
                 xytext=(points_originaux[i, 0], points_originaux[i, 1]),
                 arrowprops=dict(arrowstyle='->', color='gray', linewidth=1))


plt.xlabel('Axe X') # Label de l'axe x en fran√ßais
plt.ylabel('Axe Y') # Label de l'axe y en fran√ßais
plt.title('Transformation de Points 2D par Multiplication Matricielle') # Titre en fran√ßais
plt.legend() # Afficher la l√©gende
plt.grid(True, alpha=0.3) # Afficher la grille
plt.axhline(0, color='grey', lw=0.5) # Ligne horizontale √† y=0
plt.axvline(0, color='grey', lw=0.5) # Ligne verticale √† x=0
plt.axis('equal') # Assurer que les axes ont la m√™me √©chelle pour visualiser correctement la transformation
plt.show() # 10. Display the plot
````

````python
%%markdown
## 6. Esp√©rance, Variance et √âcart-type

En statistique, nous utilisons des mesures pour d√©crire les caract√©ristiques d'une distribution de donn√©es. Parmi les plus importantes figurent l'esp√©rance, la variance et l'√©cart-type.

*   **L'Esp√©rance (ou moyenne)**: Not√©e E(X) ou $\mu$ (mu), l'esp√©rance est la valeur moyenne attendue d'une variable al√©atoire. C'est une mesure de la **tendance centrale** de la distribution, indiquant autour de quelle valeur les donn√©es se regroupent. En science des donn√©es, la moyenne d'un √©chantillon est souvent utilis√©e comme une estimation de l'esp√©rance de la population.

*   **La Variance**: Not√©e Var(X) ou $\sigma^2$ (sigma carr√©), la variance mesure la **dispersion** ou l'√©talement des donn√©es autour de l'esp√©rance. Une variance √©lev√©e indique que les donn√©es sont tr√®s dispers√©es, tandis qu'une faible variance indique que les donn√©es sont regroup√©es pr√®s de la moyenne. Math√©matiquement, c'est la moyenne des carr√©s des √©carts par rapport √† l'esp√©rance : $Var(X) = E[(X - \mu)^2]$.

*   **L'√âcart-type**: Not√© $\sigma$ (sigma), l'√©cart-type est la racine carr√©e de la variance. Il s'agit √©galement d'une mesure de la **dispersion**, mais il est exprim√© dans les m√™mes unit√©s que les donn√©es originales, ce qui le rend plus facile √† interpr√©ter que la variance. Un √©cart-type √©lev√© signifie une grande variabilit√© des donn√©es.

En science des donn√©es et en IA :
*   Ces mesures sont utilis√©es pour **r√©sumer et comprendre** les ensembles de donn√©es.
*   Elles sont fondamentales dans de nombreux **mod√®les statistiques** et **tests d'hypoth√®ses**.
*   La variance et l'√©cart-type sont essentiels pour √©valuer la **variabilit√©** et le **risque** dans les donn√©es ou les pr√©dictions d'un mod√®le.
*   La **standardisation** des donn√©es (soustraire la moyenne et diviser par l'√©cart-type) est une √©tape de pr√©traitement courante pour de nombreux algorithmes d'apprentissage automatique.
````

````python
%%markdown
## 7. Corr√©lation Lin√©aire

La **corr√©lation lin√©aire** est une mesure statistique qui d√©crit la force et la direction d'une relation lin√©aire entre deux variables quantitatives. Le coefficient de corr√©lation de Pearson, not√© $r$, est la mesure la plus couramment utilis√©e. Sa valeur est toujours comprise entre -1 et +1 :

*   $r = +1$ : Indique une corr√©lation lin√©aire positive parfaite. Lorsque l'une des variables augmente, l'autre augmente √©galement de mani√®re parfaitement proportionnelle.
*   $r = -1$ : Indique une corr√©lation lin√©aire n√©gative parfaite. Lorsque l'une des variables augmente, l'autre diminue de mani√®re parfaitement proportionnelle.
*   $r = 0$ : Indique l'absence de corr√©lation lin√©aire. Il n'y a pas de relation lin√©aire apparente entre les deux variables.

En science des donn√©es et en IA :

*   **Analyse exploratoire des donn√©es (EDA) :** La corr√©lation est une √©tape cl√© de l'EDA pour identifier les relations potentielles entre les caract√©ristiques (features). Une forte corr√©lation entre une caract√©ristique et la variable cible peut sugg√©rer que cette caract√©ristique est un bon pr√©dicteur.
*   **S√©lection de caract√©ristiques :** Des caract√©ristiques fortement corr√©l√©es entre elles (multicollin√©arit√©) peuvent poser probl√®me dans certains mod√®les (comme la r√©gression lin√©aire). L'analyse de corr√©lation aide √† identifier ces cas.
*   **Mod√©lisation :** De nombreux mod√®les (par exemple, la r√©gression lin√©aire) se basent sur l'existence de relations (lin√©aires ou non) entre les variables.
*   **Comprendre les donn√©es :** La corr√©lation aide √† comprendre comment les variables se comportent ensemble, ce qui est crucial pour l'interpr√©tation des donn√©es.

Il est important de se rappeler que la corr√©lation n'implique pas la causalit√©. Une forte corr√©lation entre deux variables ne signifie pas n√©cessairement que l'une cause l'autre.
````

````python
import numpy as np
import matplotlib.pyplot as plt

# Generate two sets of random data with a linear relationship and some noise
np.random.seed(42) # for reproducibility

# Variable ind√©pendante (par exemple, temps pass√© √† √©tudier)
x_data = np.random.rand(100) * 10

# Variable d√©pendante avec une relation lin√©aire positive et du bruit (par exemple, note d'examen)
# note = 5 * temps_etude + bruit
y_data_positive = 5 * x_data + np.random.randn(100) * 5

# Variable d√©pendante avec une relation lin√©aire n√©gative et du bruit (par exemple, temps pass√© √† regarder la TV)
# temps_tv = -3 * temps_etude + bruit
y_data_negative = -3 * x_data + 30 + np.random.randn(100) * 5

# Variable sans relation lin√©aire (par exemple, nombre de chaussures poss√©d√©es)
y_data_none = np.random.rand(100) * 50


# Calculate the linear correlation coefficient for each pair
correlation_positive = np.corrcoef(x_data, y_data_positive)[0, 1]
correlation_negative = np.corrcoef(x_data, y_data_negative)[0, 1]
correlation_none = np.corrcoef(x_data, y_data_none)[0, 1]


# Print the calculated correlation coefficients
print(f"Coefficient de corr√©lation (x_data vs y_data_positive) : {correlation_positive:.3f}")
print(f"Coefficient de corr√©lation (x_data vs y_data_negative) : {correlation_negative:.3f}")
print(f"Coefficient de corr√©lation (x_data vs y_data_none) : {correlation_none:.3f}")


# Create scatter plots to visualize the relationships
fig, axes = plt.subplots(1, 3, figsize=(18, 6)) # 3 subplots side-by-side

# Plot 1: Positive correlation
axes[0].scatter(x_data, y_data_positive, alpha=0.7, color='skyblue')
axes[0].set_xlabel('Variable X (Temps pass√© √† √©tudier)')
axes[0].set_ylabel('Variable Y (Note d\'examen)')
axes[0].set_title(f'Corr√©lation Positive (r = {correlation_positive:.3f})')
axes[0].grid(True, alpha=0.3)

# Plot 2: Negative correlation
axes[1].scatter(x_data, y_data_negative, alpha=0.7, color='lightcoral')
axes[1].set_xlabel('Variable X (Temps pass√© √† √©tudier)')
axes[1].set_ylabel('Variable Y (Temps pass√© √† regarder la TV)')
axes[1].set_title(f'Corr√©lation N√©gative (r = {correlation_negative:.3f})')
axes[1].grid(True, alpha=0.3)

# Plot 3: No correlation
axes[2].scatter(x_data, y_data_none, alpha=0.7, color='lightgreen')
axes[2].set_xlabel('Variable X (Temps pass√© √† √©tudier)')
axes[2].set_ylabel('Variable Y (Nombre de chaussures poss√©d√©es)')
axes[2].set_title(f'Pas de Corr√©lation Lin√©aire (r = {correlation_none:.3f})')
axes[2].grid(True, alpha=0.3)

plt.tight_layout() # Adjust layout to prevent overlap
plt.show() # Display the plots
````

````python
%%markdown
## 8. Statistiques Descriptives : Moyenne, M√©diane, Max et Min

Les **statistiques descriptives** sont utilis√©es pour r√©sumer et d√©crire les caract√©ristiques principales d'un ensemble de donn√©es de mani√®re quantitative. Les mesures de tendance centrale et de dispersion sont essentielles.

*   **Moyenne (Mean)** : C'est la somme de toutes les valeurs divis√©e par le nombre de valeurs. C'est la mesure de tendance centrale la plus courante, repr√©sentant le "centre de masse" des donn√©es. Elle est sensible aux valeurs extr√™mes (outliers).

*   **M√©diane (Median)** : C'est la valeur qui se trouve exactement au milieu d'un ensemble de donn√©es ordonn√©. Si le nombre de donn√©es est pair, c'est la moyenne des deux valeurs du milieu. La m√©diane est une mesure de tendance centrale robuste aux outliers, ce qui la rend utile pour les distributions asym√©triques.

*   **Maximum (Maximum)** : C'est la valeur la plus √©lev√©e dans l'ensemble de donn√©es.

*   **Minimum (Minimum)** : C'est la valeur la plus basse dans l'ensemble de donn√©es.

En science des donn√©es et en IA :

*   Ces statistiques fournissent un **premier aper√ßu rapide** de la distribution et de la port√©e des donn√©es.
*   Elles aident √† **identifier les probl√®mes** dans les donn√©es, comme la pr√©sence d'outliers (en comparant moyenne et m√©diane) ou des plages de valeurs inattendues (via max et min).
*   Elles sont souvent utilis√©es dans les **√©tapes initiales de l'analyse exploratoire des donn√©es (EDA)**.
*   Le minimum et le maximum d√©finissent la **plage** des donn√©es, ce qui peut √™tre important pour la normalisation ou la mise √† l'√©chelle.
*   Comprendre la moyenne et la m√©diane aide √† √©valuer la **sym√©trie** ou l'**asym√©trie** (skewness) d'une distribution.
````

**Reasoning**:
Generate a dataset, calculate descriptive statistics, print the results, and create a plot to visualize the data and the calculated statistics.



````python
# 2. In a new code cell, use NumPy to generate a dataset (e.g., a random sample from a distribution, potentially with some outliers).
np.random.seed(42) # for reproducibility
# Generate data from a normal distribution
data = np.random.normal(loc=50, scale=15, size=200)

# Add some outliers
outliers = np.array([5, 10, 120, 130])
data_with_outliers = np.concatenate((data, outliers))

# 3. Calculate the mean, median, maximum, and minimum of the generated dataset using NumPy.
mean_value = np.mean(data_with_outliers)
median_value = np.median(data_with_outliers)
max_value = np.max(data_with_outliers)
min_value = np.min(data_with_outliers)

# 4. Print the calculated values with descriptive French labels.
print(f"Donn√©es g√©n√©r√©es (avec outliers), premi√®res 10 valeurs : {data_with_outliers[:10]}")
print(f"\nStatistiques Descriptives :")
print(f"  Moyenne : {mean_value:.3f}")
print(f"  M√©diane : {median_value:.3f}")
print(f"  Maximum : {max_value:.3f}")
print(f"  Minimum : {min_value:.3f}")


# 5. Use Matplotlib to create a plot that visualizes the dataset and indicates the calculated statistics.
plt.figure(figsize=(10, 6))

# Create a histogram of the data
plt.hist(data_with_outliers, bins=20, alpha=0.7, color='skyblue', edgecolor='black')

# Add vertical lines for mean and median
plt.axvline(mean_value, color='red', linestyle='dashed', linewidth=2, label=f'Moyenne ({mean_value:.3f})')
plt.axvline(median_value, color='green', linestyle='dashed', linewidth=2, label=f'M√©diane ({median_value:.3f})')

# Add horizontal lines or text for max and min
plt.text(max_value, plt.ylim()[1] * 0.95, f'Max ({max_value:.1f})', color='purple', ha='right')
plt.text(min_value, plt.ylim()[1] * 0.95, f'Min ({min_value:.1f})', color='orange', ha='left')


# 6. Add French labels to the axes and a French title to the plot.
plt.xlabel('Valeur des donn√©es') # Label de l'axe x en fran√ßais
plt.ylabel('Fr√©quence') # Label de l'axe y en fran√ßais
plt.title('Distribution des donn√©es avec Statistiques Descriptives') # Titre en fran√ßais
plt.legend() # Afficher la l√©gende
plt.grid(True, alpha=0.3) # Afficher la grille

# 7. Display the plot.
plt.show()
````

## Quartiles

Create a Markdown cell to explain quartiles and IQR in French.



````python
%%markdown
## 9. Quartiles et Intervalle Interquartile (IQR)

Les **quartiles** sont des valeurs qui divisent un ensemble de donn√©es ordonn√© en quatre parties √©gales. Ce sont des mesures de position et de dispersion qui compl√®tent la moyenne et la m√©diane. Ils sont particuli√®rement utiles pour comprendre la r√©partition des donn√©es et identifier les valeurs aberrantes (outliers).

Il existe trois quartiles principaux :

*   **Premier Quartile (Q1)** : C'est la m√©diane de la premi√®re moiti√© des donn√©es (celles qui sont inf√©rieures √† la m√©diane globale). 25% des donn√©es sont inf√©rieures ou √©gales √† Q1.
*   **Deuxi√®me Quartile (Q2)** : C'est exactement la **M√©diane** de l'ensemble des donn√©es. 50% des donn√©es sont inf√©rieures ou √©gales √† Q2.
*   **Troisi√®me Quartile (Q3)** : C'est la m√©diane de la seconde moiti√© des donn√©es (celles qui sont sup√©rieures √† la m√©diane globale). 75% des donn√©es sont inf√©rieures ou √©gales √† Q3 (ou 25% sont sup√©rieures ou √©gales √† Q3).

L'**Intervalle Interquartile (IQR)** est la diff√©rence entre le troisi√®me et le premier quartile : **IQR = Q3 - Q1**.

L'IQR repr√©sente l'√©tendue des 50% du milieu des donn√©es. C'est une mesure de **dispersion** qui est, comme la m√©diane, **robuste aux outliers**, car elle ne d√©pend pas des valeurs extr√™mes. On utilise souvent l'IQR pour identifier les outliers : une valeur est consid√©r√©e comme potentiellement aberrante si elle est inf√©rieure √† Q1 - 1.5 * IQR ou sup√©rieure √† Q3 + 1.5 * IQR.

En science des donn√©es et en IA :

*   Les quartiles et l'IQR sont essentiels pour l'**analyse exploratoire des donn√©es (EDA)** afin de comprendre rapidement la distribution et l'√©talement.
*   Ils sont utilis√©s pour **d√©tecter et g√©rer les outliers**.
*   Les diagrammes en bo√Æte (box plots), qui visualisent les quartiles et l'IQR, sont des outils courants pour comparer les distributions de diff√©rentes groupes ou variables.
*   Ils fournissent une vue plus compl√®te de la dispersion que l'√©cart-type lorsque les donn√©es ne suivent pas une distribution normale ou contiennent des outliers.
````

**Reasoning**:
Generate data with outliers, calculate quartiles and IQR, print the results, and visualize them using a histogram with vertical lines and French labels.



````python
# 4. Use NumPy to generate a dataset, potentially including some outliers.
np.random.seed(43) # Using a different seed for a new dataset
# Generate data from a slightly skewed distribution
data_quartiles = np.random.gamma(shape=5, scale=10, size=300)

# Add some outliers
outliers_quartiles = np.array([5, 10, 300, 350, 400])
data_quartiles_with_outliers = np.concatenate((data_quartiles, outliers_quartiles))


# 5. Calculate the first quartile (Q1), the second quartile (Q2 - median), and the third quartile (Q3).
q1 = np.percentile(data_quartiles_with_outliers, 25)
q2_median = np.percentile(data_quartiles_with_outliers, 50) # Q2 is the median
q3 = np.percentile(data_quartiles_with_outliers, 75)

# 6. Calculate the Interquartile Range (IQR).
iqr = q3 - q1

# 7. Print the calculated values for Q1, Q2, Q3, and IQR with descriptive French labels.
print(f"Donn√©es g√©n√©r√©es (avec outliers), premi√®res 10 valeurs : {data_quartiles_with_outliers[:10]}")
print(f"\nCalcul des Quartiles et de l'IQR :")
print(f"  Premier Quartile (Q1) : {q1:.3f}")
print(f"  Deuxi√®me Quartile (Q2 - M√©diane) : {q2_median:.3f}")
print(f"  Troisi√®me Quartile (Q3) : {q3:.3f}")
print(f"  Intervalle Interquartile (IQR = Q3 - Q1) : {iqr:.3f}")


# 8. Use Matplotlib to create a plot that visualizes the dataset and indicates the calculated quartiles.
plt.figure(figsize=(10, 6))

# Create a histogram of the data
plt.hist(data_quartiles_with_outliers, bins=30, alpha=0.7, color='lightblue', edgecolor='black')

# 9. Add vertical lines to the plot to clearly show the positions of Q1, Q2 (median), and Q3.
plt.axvline(q1, color='red', linestyle='dashed', linewidth=2, label=f'Q1 ({q1:.3f})')
plt.axvline(q2_median, color='green', linestyle='dashed', linewidth=2, label=f'Q2 (M√©diane) ({q2_median:.3f})')
plt.axvline(q3, color='purple', linestyle='dashed', linewidth=2, label=f'Q3 ({q3:.3f})')


# 10. Add French labels to the axes and a French title to the plot.
plt.xlabel('Valeur des donn√©es') # Label de l'axe x en fran√ßais
plt.ylabel('Fr√©quence') # Label de l'axe y en fran√ßais
plt.title('Distribution des donn√©es avec Quartiles et M√©diane') # Titre en fran√ßais
plt.legend() # Afficher la l√©gende
plt.grid(True, alpha=0.3) # Afficher la grille

# 11. Display the plot.
plt.show()
````
